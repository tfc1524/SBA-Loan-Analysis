```python
import pandas as pd
import numpy as np
```

```python
with open('美國州.txt', "r", encoding="utf-8") as f:
    state_list = f.readlines()
```

```python
for i in range(len(state_list)):
    state_list[i] = state_list[i].strip('\n')
```

```python
state_city = dict()

with open('美國州城市.txt', "r", encoding="utf-8") as f:
    sc_list = f.readlines()
    
    for i in range(len(sc_list)):
        sc_list[i] = sc_list[i].strip('\n')
```

```python
Now_State = 'ALABAMA'

for sc in sc_list:
    if sc in state_list:
        Now_State = sc
        state_city[Now_State] = []
    else:
        state_city[Now_State].append(sc)
```

```python
state_city_list = []

for key in state_city.keys():
    for city in state_city[key]:
        state_city_list.append(tuple([key, city]))
```
